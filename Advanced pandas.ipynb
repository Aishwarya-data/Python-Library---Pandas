## 1.GROUPBY FUNCTION:

* It is used to group data based on a column and then apply operations like: **Aggregation functions:** sum,mean,count,max / min

* Group similar values → then calculate something

```python
import pandas as pd
ins_data = pd.read_csv("insurance.csv")
ins_data
```

### Step 1 : perform initial analysis :

```python
ins_data.isna().sum()

# isna and isnull are same ---> check null values and sum () count the null values in each observations
```

**check data types**

```python
ins_data.dtypes
```

**Describes overall statistical summary**

```python
ins_data.describe()
```

**unique -- check unique values**

**nunique -- count no of unique values**

```python
ins_data["region"].unique()
```

```python
ins_data["region"].nunique()
```

### 1.pull up the average  insurance charges based on region

```python
ins_data.groupby(by = "region")["charges"].mean()
```

```python
ins_data.groupby(by = "region")["charges"].mean().sort_values(ascending = False).round(2)  # chain method
```

### 2.pull up the average  insurance charges based on region and sex

**Types of data:**

1. Discrete -- countable, it will never have units and not measurable
2. Continuous -- Measurable and have units [ unit  -  years, rs,kg,gms etc]

```python
ins_data.groupby(by = ["region","sex"])["charges"].mean().sort_values(ascending=False).round(2)
```

```python
ins_data.groupby(by = ["region","sex"])["charges"].max().sort_values(ascending=False).round(2)
```

### NOTE : FOR GROUPBY INSIDE -- DISCRETE[CATEGORICAL] ; OUTSIDE -- CONTINUOUS[NUMBERS]

## 2. PIVOT TABLE:

* A pivot table is used to summarize data by rows, columns, and aggregation.
* presentable

It helps answer questions like:

Avg salary by department?
Total sales by city?
Marks by class and gender?

```python
pd.pivot_table(data=ins_data,values="charges",index="region",columns=["sex","children"],aggfunc = 'mean').round(2)
```

### NOTE:when we want to analyse 1 or more categorical data [ discrete ] with 1 continous data [ numerical ] at that time pivot gives more presentable features -- INSIDE GROUPBY

## 3.AGGREGATE:

```python
ins_data
```

### Based on region Find average,min,max,sum of insurance charges

```python
ins_data.groupby(by="region")["charges"].aggregate(["sum","min","max","mean","count","std"]).round(2)
```

```python
ins_data.groupby(by="region")["charges"].agg(["sum","min","max","mean","count","std"]).round(2)
```

**NOTE: agg and aggregate function are same used to find average,min,max,sum etc...**

## 4.CROSS TAB

#### crosstab() in pandas is used to find frequency/count relationships between **categories**.

### How many smokers and non smokers in region wise?

```python
pd.crosstab(index = ins_data["region"],columns=ins_data["smoker"],margins=True)  

# MARGIN gives subtotals 
```

**NOTE: here we can't pass data directly so use data[column]**

**compare only 2 categorical data  -- Here -- region,smoker**

## 5.CONCAT:

concat() in pandas is used to combine multiple Series or DataFrames together.

Think of it like stacking tables

* vertically 
* horizontally 

* concat() = stacking data

* axis=0 → rows  -- **default vertical concatenation**    
* axis=1 → columns -- **horizontal concatenation**


NOTE: all columns should be same in all file

```python
sales_data_2017 = pd.read_csv("Sales Transactions-2017.csv")
sales_data_2018 =pd.read_csv("Sales Transactions-2018.csv")
sales_data_2019 =pd.read_csv("Sales Transactions-2019.csv")
```

```python
All_Sales_data = pd.concat([sales_data_2017,sales_data_2018,sales_data_2019])
```

```python
All_Sales_data
```

```python
All_Sales_data.dtypes
```

```python
All_Sales_data.isna().sum()
```

### 1.CREATE A NEW COLUMN IN THE SAME DATAFRAME WITH THE NAME "ELIGIBLITY CRITERIA" WHERE IF AGE IS > 20 SAY ELIGIBLE ELSE NOT 

```python
ins_data["age"] > 20
```

```python
def get_age(x):
    if x > 20:
        return "Eligible"
    else:
        return "Not Eligible"
```

```python
# CREATE NEW COLUMN 
```

```python
ins_data["Eligible Criteria"] = ins_data["age"].apply(get_age)
ins_data
```

NOTE:

1. **print** will just print the output but **RETURN** will impact the original data

## 7.LAMBDA FUNCTION:

* lambda function in Python is a small anonymous function.

* “anonymous” means function without a name.

* SINGLE EXPRESSION

It is heavily used with:

* pandas apply()
* map()
* sorting
* filtering
* quick transformations

```python
ins_data["Eligible Criteria_2"] = ins_data["age"].apply(lambda x:"Y" if x>20 else "N")
ins_data
```

```python
ins_data["Eligible Criteria_3"] = ins_data["age"].apply(lambda x:1 if x>20 else 0)
ins_data
```
