### 2.Give me the list of top 5 selling items

### groupby():

It is used to group data based on a column and then apply operations like:

**Aggreagation functions:** sum,mean,count,max / min

Group similar values → then calculate something

```python
order_details.groupby("item_name")["quantity"].sum()
```

```python
order_details.groupby("item_name")["quantity"].min()
```

```python
order_details.groupby("item_name")["quantity"].count()
```

```python
order_details.groupby("item_name")["quantity"].mean()
```

### Multiple aggregation

```python
order_details.groupby("item_name")["quantity"].agg(["min","max","sum","count","mean"])
```

### sort_values:

sort the values in ascending ordr by default

```python
order_details.groupby("item_name")["quantity"].sum().sort_values()   # ASCENDING
```

```python
order_details.groupby("item_name")["quantity"].sum().sort_values(ascending=False)  
```

```python
order_details.groupby("item_name")["quantity"].sum().sort_values(ascending=False).head()
```

#### NOW THIS IS THE TOP 5 SELLING ITEMS

### 1.find out top 5 selling items individual revenue it made

```python
order_details["revenue"] = order_details["item_price"]*order_details["quantity"]
order_details
```

```python
order_details.groupby("item_name")["revenue"].sum().sort_values(ascending=False).head()
```

* sort_values() → ranking

## 2.total revenue of top 5 items sold

```python
order_details.groupby("item_name")["revenue"].sum().nlargest(5).sum()
```

* groupby() → group items
* sum() → total revenue
* nlargest(5) → top 5
* sum() → final total

## step 2 Find out Null entries:

What are Null Entries?

* Null means missing data

In Pandas, null values are usually shown as:

* NaN
* None

**Find All Null Values**

```python
emp_data.isnull()
```

**Count Null Values in Each Column**

```python
emp_data.isnull().sum()
```

**Total Null Values in Dataset**

```python
emp_data.isnull().sum().sum()
```

**Remove Nulls using dropna**

**What does dropna() do?**

* Removes rows or columns that contain null (NaN) values

1. **Remove Rows with Any Null** dropna()
2. **Remove Columns with Any Null** dropna(axis=1)
3. **Remove Only if All Values Are Null** dropna(how="all")
4. **Remove Rows Based on Specific Column** dropna(subset=["Name"])
5. **inplace removal** dropna(inplace=True) -- update original data


```python
emp_data.dropna(inplace=True)
```
